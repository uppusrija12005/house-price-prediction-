```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Area': [500, 800, 1000, 1200, 1500, 1800],
    'Rooms': [1, 2, 2, 3, 3, 4],
    'Price': [1000000, 1500000, 1800000, 2200000, 2600000, 3000000]
}

df = pd.DataFrame(data)
print(df)
```

```python
X = df[['Area', 'Rooms']]   # Input
y = df['Price']             # Output
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Predicted Prices:", y_pred)
```

```python
plt.scatter(df['Area'], df['Price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

## Insights

- As area increases, house price also increases.
- Number of rooms also affects the price.
- There is a positive relationship between area and price.

## Conclusion

This project demonstrates how machine learning can be used to predict house prices based on features like area and number of rooms. Linear Regression was used to build the model and make predictions.
